```python
import re

SENTENCE_SEPARATOR = ["។", "៕", "៖", "ៗ", "?", "!"]
paragraph = 'នៅកិច្ចសន្ទនានេះ តំណាងជាច្រើនបានចាត់ទុកថា ក្នុងបរិបទដែលជម្ងឺរាតត្បាត កូវីដ ១៩ កំពុងវិវឌ្ឍន៍យ៉ាងស្មុគស្មាញលើពិភពលោក ផ្នែកកសិកម្មវៀតណាមនៅតែ ពង្រីកតួនាទីជាសសរស្តម្ភនៃខឿនសេដ្ឋកិច្ច ដោយចូលរួមវិភាគទានជិត១៥%នៃ GDP ជាតិ រក្សាបាននូវកំណើនវិជ្ជមានប្រមាណ២,៧% ក្នុងឆ្នាំ២០២០។ ក្រៅពីការធានា យ៉ាងរឹងមាំនូវសន្តិសុខស្បៀងអាហារ ចំណីអាហារសម្រាប់ប្រជាជនជិត១០០លាន នាក់នោះ ទំហំនាំចេញកសិផលក្នុងឆ្នាំទៅមិញសម្រេចបានទឹកប្រាក់ចំនួនជាង៤១,៥ ពាន់លានដុល្លាអាមេរិក។ គិតចាប់តាំងពីដើមឆ្នាំ២០២១មក ទំហំនាំចេញកសិផល វៀត ណាមក៏សម្រេចបានទឹកប្រាក់ជិត២៣ពាន់លានដុល្លាអាមេរិក ពោលគឺកើនឡើង ៣០%បើប្រៀបធៀបនឹងរយៈពេលដូចគ្នានៃឆ្នាំមុនផងដែរ។ អនុរដ្ឋមន្រ្តីក្រសួងកសិកម្ម និងអភិវឌ្ឍជនបទ វៀតណាមលោក Le Quoc Doanh បានបញ្ជាក់ថា៖ “ដោយជួយជ្រោមជ្រែងសេចក្ដីអំពាវនាវរបស់អគ្គលេខាធិការអង្គការសហ ប្រជាជាតិ វៀតណាមនឹងអនុវត្តន៍ឲ្យបានប្រសើរជាងនិងមានប្រសិទ្ធភាពជាងទៀតនូវ កម្មវិធីសកម្មភាព ដែលវិស័យនីមួយៗ ផ្នែកនីមួយៗ បានចេញផ្សាយអំពីការងារកាត់ បន្ថយភាពក្រីក្រ ធានាសន្តិសុខស្បៀងអាហារនិងអាហារូបត្ថម្ភនៅតាមតំបន់ជនបទ ជាពិសេសគឺតំបន់ឆ្ងាយដាច់ស្រយាល តំបន់ជនជាតិភាគតិច និងក្រុមមនុស្សងាយ រងគ្រោះ រួមទាំងស្ត្រីនិងកុមារផងដែរ។ ទន្ទឹមនឹងនោះ ជំរុញប្រព័ន្ធស្បៀងអាហារ ចំណី អាហារឆ្លាតវៃដែលបន្សាំនឹងការប្រែប្រួលអាកាសធាតុ ការអភិវឌ្ឍប្រកបដោយចីរភាព ធានាសន្តិសុខស្បៀងអាហារនិងអាហារូបត្ថម្ភមិនត្រឹមតែសម្រាប់ប្រជាជនវៀតណាម ជិត១០០លាននាក់ប៉ុណ្ណោះទេ ថែមទាំងក្លាយទៅជាអ្នកផ្គត់ផ្គង់ស្បៀងអាហារ ចំណី អាហារដោយមានតម្លាភាព ទំនួលខុសត្រូវ និងចីរភាពសម្រាប់ពិភពលោកទាំងមូល ផងដែរ”៕'
sentences = re.split(r"(?<=[" + "".join(SENTENCE_SEPARATOR) + "])\s*", paragraph)

count = 0
for sentence in sentences[:-1]:
  count = count + 1
  print(count, sentence)
```
